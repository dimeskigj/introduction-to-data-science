```python
import pandas as pd
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
X_test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proekt/out/t_x_test.csv")
X_train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proekt/out/t_x_train.csv")
y_test = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proekt/out/t_y_test.csv")
y_train = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proekt/out/t_y_train.csv")
```

```python
max_depths = (3, 6, 9)
learning_rates = (0.01, 0.1, 0.2, 0.3)
n_estimators = (150, 300, 650, 1000)
colsample_bytree = (0.5, 0.8, 1)
```

```python
max_depths_res = []
learning_rates_res = []
n_esimators_res = []
score_res = []
gamma_res = []
colsample_bytree_res = []
```

```python
for m_d in max_depths:
    for l_r in learning_rates:
        for n_e in n_estimators:
          for c_t in colsample_bytree:

              model = XGBRegressor(n_estimators=n_e, max_depth=m_d, learning_rate=l_r, colsample_bytree=c_t)
              model.fit(X_train, y_train)
              y_pred = model.predict(X_test)
              score = mean_squared_error(y_test, y_pred)
              
              max_depths_res.append(m_d)
              learning_rates_res.append(l_r)
              n_esimators_res.append(n_e)
              colsample_bytree_res.append(c_t)
              score_res.append(score)
```

```python
score_res = [sqrt(s) for s in score_res]
```

```python
table = pd.DataFrame(
    {
        'score': score_res,
        'max_depth': max_depths_res,
        'learning_rate': learning_rates_res,
        'n_estimators': n_esimators_res,
        'colsample_bytree': colsample_bytree_res,
    }
)
```
