```python
import pandas as pd
```

```python
X = pd.read_csv('../out/no_outlier_train_X.csv')
y = pd.read_csv('../out/no_outlier_train_Y.csv')
test = pd.read_csv('../out/kaggle_test_X.csv', index_col='Id')
```

```python
X.head()
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(n_estimators=500, learning_rate=0.01, max_depth=10)

# Add silent=True to avoid printing out updates with each cycle
model.fit(X, y)

# make predictions
y_pred = model.predict(test)
```

```python
out = pd.DataFrame(y_pred, index=test.index)
out.to_csv('sample_submission.csv')
out.head()
```

```python
for col, score in zip(X.columns, model.feature_importances_):
    print(col, score)
```
