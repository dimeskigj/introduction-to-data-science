### Колони

Ги разгледувам сите променливи и што значат, пробувам да одредам по интуиција важност на секоја колона. Секоја колона ја означувам со нејзиниот тип. 

- **име на колона** - [тип, важност (висока, средна, ниска)] опис

- **Street** - [категориски, С] Типот на улицата поврзана со имотот. 
- **Alley** - [категориски, Н] Типот на спореден пат поврзан со имотот. Исти вредности како Street, само што може и да не постои спореден пат.
- **LotShape** - [категориски, Н] Форма на имотот - регуларна, малку нерегуларна, ирегуларна...
- **LandContour** - [категориски, С] Контурите на земјата, колку е рамно земјиштето. 
- **Utilities** - [категориски, С] Комунални услуги достапни на имотот.
