```python
import pandas as pd

# Read the data
X_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proekt/out/t_x_train.csv') 
X_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proekt/out/t_x_test.csv')
Y_train = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proekt/out/t_y_train.csv') 
Y_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/proekt/out/t_y_test.csv')
```

```python
from sklearn.model_selection import GridSearchCV
import xgboost as xgb
import math

params = { 'max_depth': [3,6,10],
           'learning_rate': [0.01, 0.05, 0.1],
           'n_estimators': [100, 500, 1000],
           'colsample_bytree': [0.3, 0.7]}
xgbr = xgb.XGBRegressor()
model = GridSearchCV(estimator=xgbr, 
                   param_grid=params,
                   scoring='neg_mean_squared_error', 
                   verbose=1)
model.fit(X_train, Y_train)
print("Best parameters:", model.best_params_)
print("Lowest RMSE: ", math.sqrt(-model.best_score_))
```

```python
pred = model.best_estimator_.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error

MSE = mean_squared_error(pred, Y_test)
print(math.sqrt(MSE))
```
